```python
def parse_json(data):
    # Initialize the dataframe from loaded JSON file
    if isinstance(data, dict):
        data = pd.json_normalize(data)
        # print("Initialized dataframe")
        parse_json(data)
    else:
        # Iterate same amount of times as rows in dataframe
        for row in range(data.shape[0]):
            # base case, just column data
            if 'ColData' in data and 'Rows.Row' not in data and 'Header.ColData' not in data:
                print("Block 0")
                # normalize data frame
                col_data = pd.json_normalize(data['ColData'][row])
                # assign and print values
                acc_name = col_data['value'][0]
                acc_id   = col_data['id'][0]
                acc_val  = col_data['value'][1]
                print(f"Block 0 output: {acc_name}, {acc_id}, {acc_val}")
            # Get the ColData in dataframe that has Rows.Row and Header.ColData
            elif 'ColData' in data and 'Rows.Row' in data and 'Header.ColData' in data:
                try:
                    print("Block 1")
                    col_data = pd.json_normalize(data['ColData'][row])
                    acc_name = col_data['value'][0] 
                    acc_id   = col_data['id'][0]
                    acc_val  = col_data['value'][1]
                    print(f"Block 1 output: {acc_name}, {acc_id}, {acc_val}")
                    parse_json(data.drop(columns = ['ColData']))
                except:
                    continue
            elif 'Rows.Row' in data and 'Header.ColData' not in data:
                print("Block 2")
                row_data = pd.json_normalize(data['Rows.Row'][row])
                parse_json(row_data)
            elif 'Rows.Row' in data and 'Header.ColData' in data:
                try:
                    print("Block 3")
                    row_data  = pd.json_normalize(data['Rows.Row'][row])
                    head_data = pd.json_normalize(data['Header.ColData'][row])
                    parse_json(row_data)
                    parse_json(head_data)
                except Exception as e:
                    continue
```

```python
parse_json(rawdata)
```

```python
def parse_json(data, *field):
    # Initialize the dataframe from loaded JSON file
    if isinstance(data, dict):
        data = pd.json_normalize(data)
        # print("Initialized dataframe")
        parse_json(data)
    else:
        # Iterate same amount of times as rows in dataframe
        for row in range(data.shape[0]):
            # base case, just column data
            if 'ColData' in data and 'Rows.Row' not in data and 'Header.ColData' not in data:
                print("Block 0")
                # normalize data frame
                col_data = pd.json_normalize(data['ColData'][row])
                # assign and print values
                acc_name = col_data['value'][0]
                acc_id   = col_data['id'][0]
                acc_val  = col_data['value'][1]
                print(f"Block 0 output: {acc_name}, {acc_id}, {acc_val}")
            # Get the ColData in dataframe that has Rows.Row and Header.ColData
            elif 'ColData' in data and 'Rows.Row' in data and 'Header.ColData' in data:
                try:
                    print("Block 1")
                    col_data = pd.json_normalize(data['ColData'][row])
                    acc_name = col_data['value'][0] 
                    acc_id   = col_data['id'][0]
                    acc_val  = col_data['value'][1]
                    print(f"Block 1 output: {acc_name}, {acc_id}, {acc_val}")
                    parse_json(data.drop(columns = ['ColData']))
                except:
                    continue
            elif 'Rows.Row' in data and 'Header.ColData' not in data:
                print("Block 2")
                row_data = pd.json_normalize(data['Rows.Row'][row])
                parse_json(row_data)
            elif 'Rows.Row' in data and 'Header.ColData' in data:
                try:
                    print("Block 3")
                    row_data  = pd.json_normalize(data['Rows.Row'][row])
                    head_data = pd.json_normalize(data['Header.ColData'][row])
                    parse_json(row_data, 'Rows.Row')
                    parse_json(head_data, 'Header.ColData')
                except Exception as e:
                    continue
            else:
                for arg in field:
                    try:
                        arg_data = pd.json_normalize(data[field][row])
                    except:
                        pass
```
